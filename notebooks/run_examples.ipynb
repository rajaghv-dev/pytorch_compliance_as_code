```python
from IPython.display import display, Image as IPImage
import pathlib, time, subprocess, sys

REPO_ROOT = pathlib.Path('.').resolve()

def run_example(path, show_pngs=True):
    """Run an example script and display its output PNGs."""
    script = REPO_ROOT / path
    t0 = time.time()
    result = subprocess.run(
        [sys.executable, str(script)],
        capture_output=True, text=True,
        env={**os.environ, 'CUDA_VISIBLE_DEVICES': '', 'MPLBACKEND': 'Agg', 'TOKENIZERS_PARALLELISM': 'false'}
    )
    elapsed = time.time() - t0
    for line in result.stdout.splitlines():
        if not any(skip in line for skip in ['[Gloo]', '[torch.onnx]', 'FutureWarning', 'UserWarning', 'onnxscript']):
            print(line)
    if result.returncode != 0:
        print("STDERR:", result.stderr[-2000:])
        raise RuntimeError(f"{path} failed (exit {result.returncode})")
    print(f"Elapsed: {elapsed:.1f}s")
    if show_pngs:
        sample_dir = script.parent / 'sample_output'
        for png in sorted(sample_dir.glob('*.png')):
            print(f"  Output: {png.name}")
            display(IPImage(str(png), width=600))
```
